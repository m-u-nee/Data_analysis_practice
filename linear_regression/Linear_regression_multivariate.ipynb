```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost_function(X,y,theta):
    m = len(y)
    cost = np.sumx((X.dot(theta)-y)**2)/(2*m)
    return cost

```

```python
import numpy as np
def gradient_descent(X, y, theta, alpha, iterations, tol=0.0001):
    m = len(y)
    cost_history = np.zeros(iterations)
    
    for i in range(iterations):
        error = X.dot(theta) - y
        gradient = alpha * X.T.dot(error) / m
        theta -= gradient
        cost_history[i] = cost_function(X, y, theta)
        
        

    return theta, cost_history
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

# Load the dataset
dataset = pd.read_csv('50_startups.csv')
dataset = dataset.drop("State", axis=1)

theta = np.zeros(len(dataset.columns))
X = dataset.drop("Profit", axis = 1)
X = np.c_[np.ones(X.shape[0]), X]
X_max = X.max(axis=0)
X_scaled = X / X_max

y = dataset["Profit"]
y_scaled = y / y.max()

theta, cost_history = gradient_descent(X_scaled, y, theta, 0.01, 100000)


# Plot the cost history
plt.figure()
plt.plot(cost_history)
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost History")
plt.show()

plt.scatter(X_scaled[:,1],y)
t = np.linspace(0, 1, 10000)

r = theta[0] + t * theta[1]  
plt.plot(t, r, color="red")
plt.show()


plt.scatter(X_scaled[:,2],y)
t = np.linspace(0, 1, 1000)

r = theta[0] + t * theta[2]  
plt.plot(t, r, color="red")
plt.show()

plt.scatter(X_scaled[:,3],y)
t = np.linspace(0, 1, 1000)

r = theta[0] + t * theta[3]  
plt.plot(t, r, color="red")
plt.show()


print(theta)
print(cost_history[-1])
```
